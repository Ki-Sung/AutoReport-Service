```python
from matplotlib import font_manager, rc


rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

pd.set_option('display.max_columns', None)
```

```python
naver_df = pd.read_csv('datas/naver_query.csv')

naver_df['date'] = pd.to_datetime(naver_df['date'], format="%Y-%m-%d", errors='raise')
naver_df['day'] = naver_df['date'].dt.day
naver_df['season'] = naver_df['date'].dt.quarter

naver_df = naver_df[list(naver_df.columns[:3]) + list(naver_df.columns[-2:]) + list(naver_df.columns[3:-2])]
naver_df
```

```python
naver_df.info()
```

```python
yearly_df = naver_df.groupby('Year').agg('mean').drop(columns=['Month', 'day', 'season'])
yearly_df
```

```python
plt.figure(figsize=(20, 7))

for col in yearly_df.columns:
    sns.lineplot(data=yearly_df, x=yearly_df.index, y=col, label=col)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

```python
top = yearly_df.loc[:, np.all(yearly_df >= 29000, axis=0)]
bottom = yearly_df.loc[:, np.all(yearly_df < 10000, axis=0)]

middle_cols = list(set(yearly_df.columns) - set(top.columns) - set(bottom.columns))
middle = yearly_df[middle_cols]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 15))

for col in yearly_df.columns:
    
    if col in top.columns:
        sns.lineplot(data=top, x=top.index, y=col, label=col, ax=ax1)
    elif col in middle.columns:
        sns.lineplot(data=middle, x=middle.index, y=col, label=col, ax=ax2)
    else:
        sns.lineplot(data=bottom, x=bottom.index, y=col, label=col, ax=ax3)

ax1.legend(bbox_to_anchor=(1.0, 1))
ax1.set_ylabel(None)
ax2.legend(bbox_to_anchor=(1.0, 1))
ax2.set_ylabel(None)
ax3.legend(bbox_to_anchor=(1.0, 1))
ax3.set_ylabel(None)
        
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_yearly_df = pd.DataFrame(MinMaxScaler().fit_transform(yearly_df), columns=yearly_df.columns)
mm_yearly_df.index = yearly_df.index
mm_yearly_df
```

```python
mm_top = mm_yearly_df[top.columns]
mm_middle = mm_yearly_df[middle.columns]
mm_bottom = mm_yearly_df[bottom.columns]

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 15))

for col in yearly_df.columns:
    
    if col in mm_top.columns:
        sns.lineplot(data=mm_top, x=mm_top.index, y=col, label=col, ax=ax1)
    elif col in mm_middle.columns:
        sns.lineplot(data=mm_middle, x=mm_middle.index, y=col, label=col, ax=ax2)
    else:
        sns.lineplot(data=mm_bottom, x=mm_bottom.index, y=col, label=col, ax=ax3)
        
ax1.legend(bbox_to_anchor=(1.0, 1))
ax1.set_ylabel(None)
ax2.legend(bbox_to_anchor=(1.0, 1))
ax2.set_ylabel(None)
ax3.legend(bbox_to_anchor=(1.0, 1))
ax3.set_ylabel(None)
        
plt.show()
```

```python
yearly_percent = np.zeros((yearly_df.shape[0] - 1, yearly_df.shape[1]))
for num1 in range(yearly_df.shape[1]):
    for num2 in range(yearly_df.shape[0] - 1):
        yearly_percent[num2][num1] = ((yearly_df.values[num2+1][num1] - yearly_df.values[num2][num1]) / yearly_df.values[num2][num1]).round(2)
```

```python
yearly_percent_df = pd.DataFrame(yearly_percent, columns=yearly_df.columns)
yearly_percent_df.index = [f"{15+num}-{16+num}" for num in range(1, 6)]
yearly_percent_df
```

```python
data = yearly_percent_df.T
data['total'] = np.sum(data, axis=1).round(2)
data.sort_values('total', ascending=False, inplace=True)
data.head(10)
```

```python
plt.figure(figsize=(20, 5))

sns.barplot(data=data, x=data.index, y='total')
plt.axhline(np.mean(data, axis=0)[-1], color='black', linestyle='--', linewidth=3)

plt.xticks(rotation=30)
plt.show()
```
